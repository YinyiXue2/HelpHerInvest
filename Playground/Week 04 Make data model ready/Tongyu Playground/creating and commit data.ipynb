```python
import os, subprocess, zipfile, pandas as pd

repo_url = "https://github.com/tongyuguo/HelpHerInvest.git"
repo_dir = "HelpHerInvest"
data_dir = os.path.join(repo_dir, "Data")

# 1. clone or pull
if os.path.isdir(os.path.join(repo_dir, ".git")):
    subprocess.run(["git", "-C", repo_dir, "pull"])
else:
    subprocess.run(["git", "clone", repo_url])

# 2. create data
rows = [
    {"symbol":"AAPL","company_name":"Apple Inc.","sector":"Technology","industry":"Consumer Electronics","exchange":"NASDAQ"},
    {"symbol":"MSFT","company_name":"Microsoft Corp.","sector":"Technology","industry":"Software—Infrastructure","exchange":"NASDAQ"},
    {"symbol":"XOM","company_name":"Exxon Mobil Corp.","sector":"Energy","industry":"Oil & Gas Integrated","exchange":"NYSE"},
]

df = pd.DataFrame(rows)
os.makedirs(data_dir, exist_ok=True)

csv_path = os.path.join(data_dir, "another_test.csv")
zip_path = csv_path + ".zip"

df.to_csv(csv_path, index=False)
with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as z:
    z.write(csv_path, arcname="another_test.csv")

print("Saved:", zip_path)

# 3. check push permission
result = subprocess.run(
    ["git", "-C", repo_dir, "push", "--dry-run"],
    capture_output=True,
    text=True
)

is_collaborator = result.returncode == 0
print("Collaborator:", is_collaborator)

# 4. commit + push only if collaborator
if is_collaborator:
    subprocess.run(["git", "-C", repo_dir, "add", "Data/another_test.csv.zip"])
    subprocess.run(["git", "-C", repo_dir, "commit", "-m", "add new dataset"])
    subprocess.run(["git", "-C", repo_dir, "push"])
else:
    print("⚠️ Not a collaborator — skipping push.")
```
