```python
import os, subprocess
from pathlib import Path

REPO_URL = "https://github.com/tongyuguo/HelpHerInvest.git"
REPO_NAME = "HelpHerInvest"

def git_toplevel(path="."):
    r = subprocess.run(["git", "-C", path, "rev-parse", "--show-toplevel"],
                       capture_output=True, text=True)
    return r.stdout.strip() if r.returncode == 0 else None

def clone_or_pull():
    top = git_toplevel(".")
    if top and Path(top).name == REPO_NAME:
        # We are already inside the HelpHerInvest repo
        subprocess.run(["git", "pull"], check=False)
        return top

    # Otherwise, clone/pull into a sibling folder (NOT inside the repo)
    base = Path.cwd()
    repo_dir = base / REPO_NAME
    if (repo_dir / ".git").exists():
        subprocess.run(["git", "-C", str(repo_dir), "pull"], check=False)
    else:
        subprocess.run(["git", "clone", REPO_URL, str(repo_dir)], check=False)
    return str(repo_dir)

REPO_DIR = clone_or_pull()
print("Using repo at:", REPO_DIR)
```

```python
def can_push():
    result = subprocess.run(
        ["git", "-C", REPO_DIR, "push", "--dry-run"],
        capture_output=True, text=True
    )
    return result.returncode == 0

def commit_and_push(msg="update"):
    subprocess.run(["git", "-C", REPO_DIR, "add", "-A"])
    subprocess.run(["git", "-C", REPO_DIR, "commit", "-m", msg])
    subprocess.run(["git", "-C", REPO_DIR, "push"])

if can_push():
    commit_and_push("update notebook / data")
else:
    print("Not a collaborator — skipping push")
```
