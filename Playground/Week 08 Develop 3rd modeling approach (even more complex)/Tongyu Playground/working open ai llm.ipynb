```python
from openai import OpenAI
import zipfile
import os

client = OpenAI()

# Extract zip
zip_path = "./HelpHerInvest/Data/final_dataset_20260224v2.csv.zip"
extract_dir = "./HelpHerInvest/Data/"

with zipfile.ZipFile(zip_path, "r") as z:
    z.extractall(extract_dir)

csv2_path = os.path.join(extract_dir, "final_dataset_20260224v2.csv")

# Upload both files
with open("./HelpHerInvest/Data/nlp_clean_stock_symbols.csv", "rb") as f1, \
     open(csv2_path, "rb") as f2:

    file1 = client.files.create(file=f1, purpose="user_data")
    file2 = client.files.create(file=f2, purpose="user_data")

user_input = "I like the fashion industry."

response = client.responses.create(
    model="gpt-5-mini",
    input=[
        {
            "role": "developer",
            "content": [
                {
                    "type": "input_text",
                    "text": (
                        "You are a data scientist specializing in stock market analysis.\n"
                        "For every request, you must use the two files attached in the user message.\n"
                        "The first file is a stock symbol reference dataset.\n"
                        "The second file is a company/fundamental dataset.\n\n"
                        "Required workflow:\n"
                        "1. Analyze the user's interest.\n"
                        "2. Use the first file to identify 20 relevant stock symbols.\n"
                        "3. Use the second file to rank those 20 symbols, with machine learning algorithms such as linear regression or random forest.\n"
                        "4. Provide a brief explanation for why the stocks were selected.\n"
                        "5. Provide a justification for the ranking.\n"
                        "6. If the data or user input is insufficient, clearly say so instead of guessing."
                    )
                }
            ]
        },
        {
            "role": "user",
            "content": [
                {"type": "input_file", "file_id": file1.id},
                {"type": "input_file", "file_id": file2.id},
                {"type": "input_text", "text": user_input}
            ]
        }
    ]
)

print(response.output_text)
```
