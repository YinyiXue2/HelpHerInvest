```python
import random
import pandas as pd

random.seed(34)  # For reproducibility

# Create topics for templates
topics = {
    "technology": ["AI", "cloud computing", "datacenters", "cybersecurity", "semiconductors","quantum computing","blockchain","telecommunications","software","hardware","self-driving cars","automation"],
    "consumer": ["fashion", "gaming", "e-commerce", "beauty","home goods","food","beverage","travel","leisure","sports","automotive","retail","consumer electronics"],
    "energy": ["renewable energy", "oil and gas", "EV infrastructure", "nuclear power","energy storage","solar power","wind energy","utilities","clean energy"],
    "healthcare": ["biotech", "pharmaceuticals", "medical devices","telemedicine","health IT","genomics","healthcare services","wellness","fitness","mental health"],
    "media": ["music streaming", "broadcasting", "digital advertising","marketing","social media","entertainment","publishing","gaming media","video streaming","podcasting"],
    "finance": ["fintech", "cryptocurrency", "banking", "insurance","wealth management","investment platforms","personal finance","lending"],
}

# Generate templates
single_templates = [
    "I want to invest in {topic}.",
    "I'm interested in companies focused on {topic}.",
    "Looking for exposure to {topic}.",
    "I believe {topic} will grow significantly.",
    "I'm passionate about {topic} and want investment opportunities."
]

dual_templates = [
    "I'm interested in both {topic1} and {topic2}.",
    "I want exposure to companies in {topic1} and {topic2}.",
    "Looking to invest in {topic1} as well as {topic2}.",
    "I believe {topic1} and {topic2} will grow over the next decade.",
]

# Add modifier / adjectives to be more discriptive
modifiers = [
    "",
    "United States-based",
    "emerging",
    "innovative",
    "high-growth",
    "sustainable",
    "global leaders in",
    "disruptive",
    "cutting-edge",
    "socially responsible"
]

# Generate the sample text inputs
def generate_synthetic_inputs(n=500, dual_prob=0.35):
    rows = []
    categories = list(topics.keys())

    for _ in range(n):

        # randomize the whether there is a 1 or 2 topics
        is_dual = random.random() < dual_prob

        # if 1 topic
        if not is_dual:
            category = random.choice(categories)
            topic = random.choice(topics[category])
            template = random.choice(single_templates)

            text = template.format(topic=topic)

            rows.append({
                "primary_topic": topic,
                "secondary_topic": None,
                "is_dual": 0,
                "text_input": text,                
            })

        # else 2 topics
        else:
            # pick two different cateogries for topics
            t1, t2 = random.sample(categories, 2)

            topic1 = random.choice(topics[t1])
            topic2 = random.choice(topics[t2])

            template = random.choice(dual_templates)
            text = template.format(topic1=topic1, topic2=topic2)

            rows.append({
                "primary_topic": topic1,
                "secondary_topic": topic2,
                "is_dual": 1,
                "text_input": text               
            })

    return pd.DataFrame(rows)

df_text_inputs = generate_synthetic_inputs(1000, dual_prob=0.4)

df_text_inputs.head(20)
```
